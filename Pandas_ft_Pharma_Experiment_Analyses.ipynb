```python
import pandas as pd
data=pd.read_csv('Pfizer_1.csv')
data.head()
```

```python
data.shape
```

```python
#Convert wide to long through -MELT
#LONG TO WIDE through -PIVOT
data_melt=pd.melt(data,id_vars=['Date','Drug_Name','Parameter'],var_name='time',value_name='readings')
data_melt.head()
```

```python
data_tidy = data_melt.pivot(
    index=['Date', 'Drug_Name', 'Parameter'],
    columns='time',
    values='readings'
).reset_index()
data_tidy.head()
```

```python
data_tidy = data_melt.pivot(
    index=['Date', 'Drug_Name', 'time'],
    columns='Parameter',
    values='readings'
).reset_index()
data_tidy.head()
```

```python
data_tidy.shape
```

```python
#Filling the missing value with the respective DRUG means#data.isna().sum() To ckeck missing data
```

```python
def temp(x):
  x['temp_avg']=x['Temperature'].mean()
  return x

data_tidy=data_tidy.groupby('Drug_Name').apply(temp)
data_tidy.head()
```

```python
data_tidy['Temperature'].fillna(data_tidy['temp_avg'],inplace=True)
data_tidy.head()
```

```python
data_tidy.isna().sum()
```

```python
# Reset the index to avoid ambiguity with 'Drug_Name'
data_tidy = data_tidy.reset_index(drop=True)

# Calculate the mean pressure for each drug
pressure_avg_by_drug = data_tidy.groupby('Drug_Name')['Pressure'].mean().reset_index()
pressure_avg_by_drug.rename(columns={'Pressure': 'pressure_avg'}, inplace=True)

# Merge the average pressure back into the main DataFrame
data_tidy = data_tidy.merge(pressure_avg_by_drug, on='Drug_Name', how='left')

data_tidy.head()
```

```python
temp_points=[5,20,35,50,60]
temp_labels=['low','medium','high','very high']
data_tidy['temp_category']=pd.cut(data_tidy['Temperature'],bins=temp_points,labels=temp_labels)
data_tidy.head()
```
