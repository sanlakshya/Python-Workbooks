```python
import pandas as pd
movies=pd.read_csv('movies.csv',index_col=0)
movies.head()# Review DataSet
```

```python
directors=pd.read_csv('directors.csv',index_col=0)
directors.head()
```

```python
data=movies.merge(directors,left_on='director_id',right_on='id',how='left')
data.head()
```

```python
data.drop(['director_id','id_y'],axis=1,inplace=True)
data.head()#Removing the unnecessary columns ,both columns are same
```

```python
#Converting Gender column into boolian column
```

```python
def encode(data):
  if data=='Female':
    return 0
  else:
    return 1
data['gender']=data['gender'].apply(encode)
data.head()
```

```python
#creating a Profit Column from revenue and budget columns
```

```python
def profit(data):
  return data['revenue']-data['budget']
data['profit']=data.apply(profit,axis=1)
data.head()
```

```python
data_dir_budget=data.groupby('director_name')['budget'].max().reset_index()
data_dir_budget
```

```python
#679 are the numbers of high budgeted movies
```

```python
def func(x):
    return x['budget'].max() >= 100

data.groupby('director_name').filter(func)
```

```python
#Filter out the RISKY Movies
```

```python
def func(x):
    x['risky'] = (x['budget'] - x['revenue'].mean()) >= 0
    return x

data_risky=data.groupby('director_name').apply(func)
```

```python
data_risky.loc[data_risky['risky']==True]
```

```python
#Find the MOST Productive Movies
```

```python
data_agg=data.groupby('director_name').agg({
    'year': ['min', 'max'],
    'title': 'count'
}).reset_index()
data_agg.head()
#MULTIPLE LEVEL INDEXING ACCESS
```

```python
#Multi to single level indexing
```

```python
# Step 1: Aggregate
data_agg = data.groupby('director_name').agg({
    'year': ['min', 'max'],
    'title': 'count'
})

# Step 2: Flatten column names
data_agg.columns = ['year_min', 'year_max', 'title_count']

# Step 3: Reset index if needed (optional)
data_agg = data_agg.reset_index()

# Step 4: View the result
print(data_agg.head())
```
